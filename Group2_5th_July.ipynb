#Objective
Given a csv file containing 2016 marathon results, the objective is to use SQL to answer questions that we come up with


```python
#importing pandas module
import pandas as pd
```

```python
# loading the csv data into a dataframe
df=pd.read_csv("/marathon_results_2016.csv")
```

```python
df.head()
```

```python
#importing sqlite module
import sqlite3
```

```python
#creating a database connection
conn=sqlite3.connect("database.db")
```

```python
df.to_sql("df",conn,if_exists="append",index=False)
```

```python
# reading pd table
pd.read_sql(""" SELECT * FROM df limit 5""",conn)
```

```python
#checking tables in the database db
sql_query="""select name from sqlite_master where type='table' """
```

```python
cursor=conn.cursor()
```

```python
cursor.execute(sql_query)
```

```python
#printing tables
print(cursor.fetchall())
```

```python
#table df info
pd.read_sql("""PRAGMA table_info(df)""", conn, index_col="cid")
```
